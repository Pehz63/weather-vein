```python
import torch
import gpytorch
import numpy as np
import matplotlib.pyplot as plt
```

```python
torch.manual_seed(1)
def f(x):
    return torch.sin(2 * x)

# Training inputs
train_x = torch.linspace(0, 3, 30)
# True function values (noise-free)
true_y = f(train_x)
# Observed training targets = true function + noise
noise_std = 0.0
train_y = true_y + noise_std * torch.randn_like(train_x)
```

```python
train_x.shape, train_y.shape
```

```python
class ExactGPModel(gpytorch.models.ExactGP):
    def __init__(self, train_x, train_y, likelihood):
        super().__init__(train_x, train_y, likelihood)
        
        self.mean_module = gpytorch.means.ZeroMean()
        
        self.covar_module = gpytorch.kernels.ScaleKernel(
            gpytorch.kernels.RBFKernel()
        ) # RBF Kernel

    def forward(self, x):
        mean = self.mean_module(x)
        covar = self.covar_module(x)
        return gpytorch.distributions.MultivariateNormal(mean, covar)
```

```python
likelihood = gpytorch.likelihoods.GaussianLikelihood()
model = ExactGPModel(train_x, train_y, likelihood)
```

```python
test_x = torch.linspace(-1, 4, 100)  # a bit beyond training range

model.eval()

with torch.no_grad():
    # Prior distribution at test_x
    prior_dist = model(test_x)  # prior because we have not called model.train() or optimized
    # Sample 3 random functions from the prior
    prior_samples = prior_dist.rsample(sample_shape=torch.Size([3]))
    
plt.figure(figsize=(8,4))
for i in range(3):
    plt.plot(test_x.numpy(), prior_samples[i].numpy(), label=f'Prior sample {i+1}')
plt.title("Samples from the GP Prior")
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.show()
```

```python
model.train()
likelihood.train()

optimizer = torch.optim.Adam([
    {'params': model.parameters()}
], lr=0.1)

# "Loss" for GPs in GPyTorch is the negative marginal log-likelihood
mll = gpytorch.mlls.ExactMarginalLogLikelihood(likelihood, model)

training_iter = 1000
for i in range(training_iter):
    optimizer.zero_grad()
    output = model(train_x)
    loss = -mll(output, train_y)  #  maximize MLL => minimize negative MLL
    loss.backward()
    optimizer.step()

    if (i+1) % 20 == 0:
        print(f"Iter {i+1}/{training_iter} - Loss: {loss.item():.3f}  "
              f"lengthscale: {model.covar_module.base_kernel.lengthscale.item():.3f}  "
              f"noise: {model.likelihood.noise.item():.3f}")
```

```python
model.eval()
likelihood.eval()

with torch.no_grad():
    pred_dist = likelihood(model(test_x))
    mean = pred_dist.mean
    lower, upper = pred_dist.confidence_region()

plt.figure(figsize=(8,4))
plt.scatter(train_x.numpy(), train_y.numpy(), color='k', s=20, alpha=0.5, label='Training data')

# Ground truth (noise-free)
plt.plot(test_x.numpy(), f(test_x).numpy(), 'g--', label='True function')

# Posterior mean
plt.plot(test_x.numpy(), mean.numpy(), 'b', label='Posterior mean')

# Confidence region
plt.fill_between(test_x.numpy(),
                 lower.numpy(),
                 upper.numpy(),
                 alpha=0.2,
                 color='blue')

plt.title('GP Posterior Mean and Confidence')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.show()
```

```python
model = ExactGPModel(train_x, train_y, likelihood)
optimizer = torch.optim.Adam(model.parameters(), lr=0.1)
mll = gpytorch.mlls.ExactMarginalLogLikelihood(likelihood, model)

model.train()
likelihood.train()

loss_history = []

for i in range(training_iter):
    optimizer.zero_grad()
    output = model(train_x)
    loss = -mll(output, train_y)
    loss.backward()
    optimizer.step()
    loss_history.append(loss.item())

plt.figure(figsize=(6,4))
plt.plot(loss_history, label='Neg. Log Marginal Likelihood')
plt.xlabel('Iteration')
plt.ylabel('Loss')
plt.title('Training Progress (GP Marginal Likelihood)')
plt.legend()
plt.show()
```

## effect of different lengthscales

```python
lengthscales = [0.01, 0.1, 1.0, 10.0]

plt.figure(figsize=(10, 6))
colors = ['b', 'g', 'r', 'm']

model.eval()

for ls, c in zip(lengthscales, colors):
    # Manually set the RBF kernel lengthscale
    model.covar_module.base_kernel.lengthscale = ls
    
    # Get the PRIOR distribution (prior=True) at test_x
    with torch.no_grad():
        prior_output = model(test_x)
        # Sample a single function draw (or multiple) from the prior
        prior_sample = prior_output.sample(sample_shape=torch.Size([1]))[0]

    # Plot the single prior sample
    plt.plot(test_x.numpy(), prior_sample.numpy(), c=c, label=f'lengthscale={ls}')

plt.title("Effect of lengthscale on prior samples")
plt.legend()
plt.show()
```
