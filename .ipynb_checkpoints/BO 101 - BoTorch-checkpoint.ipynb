### imports

```python
# This code is built on top of three main python libraries: pytorch, gpytorch, and botorch
```

```python
import torch
from botorch import fit_gpytorch_model
from botorch.models import SingleTaskGP
from botorch.test_functions import Hartmann
from gpytorch.mlls import ExactMarginalLogLikelihood
from botorch.acquisition import ExpectedImprovement
from botorch.optim import optimize_acqf
```

```python
# this is how the full bayesian optimization (BO) procedure look like, 
# we will go over each step of it one by one
```

![bo_loop](bo_loop.png)

### defining the objective function 

```python
# We are using the Hartmann function as the test function for illustration purposes. 
# In the real world, this would be replaced by the actual function we want to optimize.
# We will not have closed form expressions for the function and its gradient in the real world
# and the function will be expensive to evaluate.

neg_hartmann6 = Hartmann(dim=6, negate=True)
```

![Hartmann](hartmann.png)

### setting up initial points for evaluation 

```python
# we want some initial data to start with 
NUM_STARTING_POINTS = 10
DIM = 6
train_x = torch.rand(NUM_STARTING_POINTS, DIM) # other choices : sobol 
train_obj = neg_hartmann6(train_x).unsqueeze(-1)
```

### fitting the surrogate model

```python
??SingleTaskGP # ?? is a magic command (that works in Jupyter notebook) that shows the source code of the function
```

```python
# torch.set_default_dtype(torch.float64)
```

```python
# fit a Gaussian Process (GP) model to the data 
# GP is the probabilistic model in the figure above
model = SingleTaskGP(train_X=train_x, train_Y=train_obj)
mll = ExactMarginalLogLikelihood(model.likelihood, model)
fit_gpytorch_model(mll)
```

```python
# looking at the parameters of the model
for n, v in model.named_parameters():
    print(n, v)
```

### optimizing the acquisition function

```python
# acquisition function is a function that tells us how good a point is to evaluate next
best_value = train_obj.max()
EI = ExpectedImprovement(model=model, best_f=best_value)
```

```python
??ExpectedImprovement
```

```python
# optimizing the acquisition function to find the next point to evaluate
new_point_analytic, _ = optimize_acqf(
    acq_function=EI,
    bounds=torch.tensor([[0.0] * 6, [1.0] * 6]),
    q=1,
    num_restarts=20,
    raw_samples=100,
    options={},
)
```

```python
train_x = torch.cat([train_x, new_point_analytic])
train_obj = neg_hartmann6(train_x).unsqueeze(-1)
```

## full Bayesian optimization (BO) Loop

```python
from botorch.models.transforms import Normalize, Standardize
```

```python
for _ in range(50):
    model = SingleTaskGP(train_X=train_x, train_Y=train_obj, input_transform=Normalize(d=6), outcome_transform=Standardize(m=1))
    mll = ExactMarginalLogLikelihood(model.likelihood, model)
    fit_gpytorch_model(mll)
    best_value = train_obj.max()
    EI = ExpectedImprovement(model=model, best_f=best_value)
    new_point_analytic, _ = optimize_acqf(
        acq_function=EI,
        bounds=torch.tensor([[0.0] * 6, [1.0] * 6]),
        q=1,
        num_restarts=20,
        raw_samples=100,
        options={},
    )
    train_x = torch.cat([train_x, new_point_analytic])
    train_obj = neg_hartmann6(train_x).unsqueeze(-1)
    print(f'Best value achieved after {len(train_x) -NUM_STARTING_POINTS} BO steps: {torch.max(train_obj).item()}')
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.plot(train_obj.numpy(), label='actual observed values')
plt.plot(np.maximum.accumulate(train_obj.numpy()), label='cumulative max')
plt.legend()
```
